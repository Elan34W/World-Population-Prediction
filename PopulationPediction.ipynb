```python
import pandas as pd
import matplotlib.pyplot as plt
from modsim import *
import numpy as np
from collections import namedtuple
```

## DATA POPULASI DUNIA

```python
# Membaca data dari file CSV
data_populasi = pd.read_csv("world-population-by-country-2020.csv")
data_populasi.rename(columns={"no": "No", "Country (or dependency)": "Country","Fert. Rate":"Fertility Rate"}, inplace=True)
selected_columns = data_populasi[["Country", "Population 2020", "Yearly Change", "Migrants (net)", "Land Area (Km²)", "Density (P/Km²)", "Fertility Rate"]]
```

```python
# Cek 24 baris pertama untuk memahami struktur data
selected_columns.head(20)
```

```python
# Cek informasi dataset untuk mengetahui ada tidaknya missing values
print(selected_columns.info())
```

```python
# Definisi namedtuple untuk State dan System
State = namedtuple('State', ['population', 'yearly_change', 'migrants'])
System = namedtuple('System', ['land_area', 'density', 'fertility_rate'])
```

```python
# Fungsi untuk membuat sistem
def make_system(country, density=None):
    """Membuat obyek system untuk model populasi."""
    index = selected_columns[selected_columns['Country'] == country].index[0]
    if density is None:
        density = int(selected_columns['Density (P/Km²)'][index].replace(',', ''))
    land_area = int(selected_columns['Land Area (Km²)'][index].replace(',', ''))
    fertility_rate = float(selected_columns['Fertility Rate'][index])
    population = int(selected_columns['Population 2020'][index].replace(',', ''))
    yearly_change = float(selected_columns['Yearly Change'][index].strip('%')) / 100
    migrants = int(selected_columns['Migrants (net)'][index].replace(',', ''))

    system = System(land_area, density, fertility_rate)
    state = State(population, yearly_change, migrants)
    
    return system, state
```

```python
# Fungsi untuk memperbarui state dengan mempertimbangkan land area dan density
def update_func(state, t, system):
    """Fungsi untuk memperbarui populasi pada setiap langkah waktu."""
    births = system.fertility_rate * state.population / 1000
    net_migration = state.migrants / 1000
    population_change = births + net_migration + (state.yearly_change * state.population)
    
    # Pengaruh dari luas daratan dan kepadatan penduduk
    land_effect = system.land_area / 1_000_000  # Luas daratan dalam juta km²
    density_effect = (1 - (state.population / (system.density * system.land_area))) * 0.1
    
    # Modifikasi laju pertumbuhan populasi berdasarkan luas daratan dan kepadatan penduduk
    population_change *= (1 + land_effect + density_effect)
    
    new_population = state.population + population_change
    return State(new_population, state.yearly_change, state.migrants)
```

```python
# Fungsi untuk menjalankan simulasi populasi
def run_population_simulation(system, state, years):
    """Menjalankan simulasi populasi."""
    results = pd.Series(dtype=float)
    results[0] = state.population
    
    for t in range(1, years + 1):
        state = update_func(state, t, system)
        results[t] = state.population
    
    return results
```

```python
# Dictionary untuk menyimpan semua hasil simulasi
all_results = {}

# Lama simulasi dalam tahun
years = 6

# Menjalankan simulasi untuk semua negara
for country in selected_columns['Country']:
    system, state = make_system(country)
    results = run_population_simulation(system, state, years)
    all_results[country] = results

# Plot hasil simulasi untuk beberapa negara sebagai contoh
plt.figure(figsize=(15, 10))

# Plot untuk beberapa negara pilihan agar visualisasi lebih jelas
countries_to_plot = ['Indonesia', 'Russia', 'United States', 'Brazil', 'Nigeria', 'Philippines', 'Japan', 'Mexico', 'Ethiopia', 'Egypt']

for country in countries_to_plot:
    results = all_results[country]
    plt.plot(results.index, results, marker='o', label=country)

plt.title(f'Simulasi Populasi Negara selama {years} Tahun dengan Pengaruh Luas Daratan dan Kepadatan')
plt.xlabel('Tahun')
plt.ylabel('Populasi (Ratusan Juta)')
plt.legend(loc='upper right', bbox_to_anchor=(1.15, 1))
plt.grid(True)
plt.show()

# # Cetak hasil simulasi untuk semua negara
# for country, results in all_results.items():
#     print(f"Hasil simulasi untuk {country}:")
#     print(results)
#     print()
```

```python
# Agregasi hasil simulasi untuk semua negara
total_population = pd.Series(dtype=float)

for country, results in all_results.items():
    if total_population.empty:
        total_population = results.copy()
    else:
        total_population += results

# Plot hasil agregasi
plt.figure(figsize=(15, 10))
plt.plot(total_population.index, total_population, marker='o', label='Total Populasi Dunia')
plt.title(f'Simulasi Total Populasi Dunia selama {years} Tahun dengan Pengaruh Luas Daratan dan Kepadatan')
plt.xlabel('Tahun')
plt.ylabel('Total Populasi')
plt.legend()
plt.grid(True)
plt.show()

# Cetak hasil total populasi
total_population_normal = total_population.apply(lambda x: '{:.0f}'.format(x))
print("Hasil total populasi dunia dari simulasi:")
print(total_population_normal)
```

## SIMULASI MENGGUNAKAN SWEEPSERIES

### Analisi Pengaruh Density terhadap pertumbuhan populasi di dunia

```python
# Definisi sweep series untuk faktor density
density_sweep = np.linspace(100, 1000, 10)

# Simpan hasil simulasi untuk setiap nilai density dalam satu SweepSeries
total_population_sweep = SweepSeries()

for density in density_sweep:
    # Menjalankan simulasi untuk semua negara
    all_results = {}
    for country in selected_columns['Country']:
        system, state = make_system(country, density)
        results = run_population_simulation(system, state, years)
        all_results[country] = results
    
    # Agregasi hasil simulasi untuk semua negara
    total_population = pd.Series(dtype=float)
    for country, results in all_results.items():
        if total_population.empty:
            total_population = results.copy()
        else:
            total_population += results
    
    # Simpan hasil populasi dunia ke dalam SweepSeries
    total_population_sweep[density] = total_population

# Plot hasil SweepSeries
plt.figure(figsize=(15, 10))
for density, total_population in total_population_sweep.items():
    plt.plot(total_population.index, total_population, marker='o', label=f'Density: {density:.2f}')

plt.title(f'Simulasi Total Populasi Dunia selama {years} Tahun dengan Berbagai Density')
plt.xlabel('Tahun')
plt.ylabel('Populasi')
plt.legend(loc='upper right', bbox_to_anchor=(1.16, 1))
plt.grid(True)
plt.show()
```

```python
# Buat list untuk menyimpan rata-rata populasi dunia untuk setiap tahun
average_population = []

# Iterasi melalui setiap tahun dalam hasil simulasi
for year in total_population_sweep[density_sweep[0]].index:
    # Hitung rata-rata populasi dunia untuk tahun tersebut
    total_population_yearly = 0
    for density in density_sweep:
        total_population_yearly += total_population_sweep[density][year]
    average_population.append(total_population_yearly / len(density_sweep))

# Plot rata-rata populasi dunia
plt.figure(figsize=(15, 10))
plt.plot(total_population_sweep[density_sweep[0]].index, average_population, marker='o', label='Rata-rata Populasi Dunia')
plt.title(f'Rata-rata Populasi Dunia selama {years} Tahun dengan Berbagai Density')
plt.xlabel('Tahun')
plt.ylabel('Populasi')
plt.legend(loc='upper right', bbox_to_anchor=(1.21, 1))
plt.grid(True)
plt.show()

# Cetak hasil rata-rata populasi dunia
average_population_normal = [int(pop) for pop in average_population]
print("Hasil rata-rata populasi dunia dari simulasi:")
for year, avg_pop in enumerate(average_population_normal):
    print(f"Tahun {year}: {avg_pop}")
```

```python
# Buat sweep series untuk menyimpan tingkat pertumbuhan populasi relatif terhadap density
growth_rate_sweep = SweepSeries()

# Iterasi melalui setiap nilai density dalam density_sweep
for density, total_population in total_population_sweep.items():
    # Hitung pertumbuhan populasi relatif terhadap density
    growth_rate = ((total_population[years] - total_population[0]) / total_population[0]) * 100
    
    # Simpan hasil pertumbuhan populasi relatif ke dalam sweep series
    growth_rate_sweep[density] = growth_rate

# Plot hasil sweep series
plt.figure(figsize=(15, 10))
plt.plot(growth_rate_sweep.index, growth_rate_sweep.values, marker='o')
plt.title('Pengaruh Density terhadap Tingkat Pertumbuhan Populasi')
plt.xlabel('Density')
plt.ylabel('Tingkat Pertumbuhan Populasi (%)')
plt.grid(True)
plt.show()

# Cetak hasil tingkat pertumbuhan populasi relatif terhadap density
print("Hasil tingkat pertumbuhan populasi relatif terhadap density:")
for density, growth_rate in growth_rate_sweep.items():
    print(f"Density {density:.2f}: {growth_rate:.2f}%")
```
